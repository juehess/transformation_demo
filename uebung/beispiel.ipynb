```python
import numpy as np
from matplotlib import pyplot as plt
from pytransform3d.transform_manager import TransformManager
```

```python
def plot_frames_in(frame):
    tm.plot_frames_in(frame)

    man = plt.get_current_fig_manager()
    man.set_window_title("Example")
    plt.show()
```

```python
base_frame = "base"

tm = TransformManager()
tm.add_transform(base_frame, base_frame, np.eye(4))
plot_frames_in(base_frame)
```
