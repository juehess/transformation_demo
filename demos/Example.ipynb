```python
import numpy as np
import matplotlib.pyplot as plt

# Define the transformation parameters
t_x = 8
t_y = 5
theta = 180  # in degrees

# Convert theta to radians
theta_rad = np.deg2rad(theta)

# Define the homogeneous transformation matrix from A to B
T_AB = np.array([
    [np.cos(theta_rad), -np.sin(theta_rad), t_x],
    [np.sin(theta_rad), np.cos(theta_rad), t_y],
    [0, 0, 1]
])

# Define the original coordinate axes for frame A
origin = np.array([[0, 0, 1]])
x_axis = np.array([[2, 0, 1]])
y_axis = np.array([[0, 2, 1]])

# Transform these axes to get the axes for frame B
transformed_origin = np.dot(origin, T_AB.T)
transformed_x_axis = np.dot(x_axis, T_AB.T)
transformed_y_axis = np.dot(y_axis, T_AB.T)

def plot_transformed_system_with_matching_axis_labels():
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Arrow style
    arrow_style = dict(angles='xy', scale_units='xy', scale=1, width=0.01, headwidth=3, headlength=3, headaxislength=2)
    
    # Plot original coordinate system using quiver
    ax.quiver(origin[0, 0], origin[0, 1], x_axis[0, 0]-origin[0, 0], x_axis[0, 1]-origin[0, 1],
              color='red', **arrow_style)
    ax.quiver(origin[0, 0], origin[0, 1], y_axis[0, 0]-origin[0, 0], y_axis[0, 1]-origin[0, 1],
              color='green', **arrow_style)

    # Plot transformed coordinate system using quiver
    ax.quiver(transformed_origin[0, 0], transformed_origin[0, 1], 
              transformed_x_axis[0, 0]-transformed_origin[0, 0], transformed_x_axis[0, 1]-transformed_origin[0, 1], 
              color='red', **arrow_style)
    ax.quiver(transformed_origin[0, 0], transformed_origin[0, 1], 
              transformed_y_axis[0, 0]-transformed_origin[0, 0], transformed_y_axis[0, 1]-transformed_origin[0, 1], 
              color='green', **arrow_style)

    # Add labels with larger and bold font
    ax.text(origin[0, 0]-0.3, origin[0, 1]-0.3, "A", fontsize=18, color='black', ha='center', va='center', fontweight='bold')
    ax.text(transformed_origin[0, 0]+0.3, transformed_origin[0, 1]-0.3, "B", fontsize=18, color='black', ha='center', va='center', fontweight='bold')
    
    # Uniform position for axis labels
    x_label_offset, y_label_offset = 0.2, 0.2
    
    # Axis labels for original coordinate system
    ax.text(x_axis[0, 0]+x_label_offset, x_axis[0, 1]-y_label_offset, "x", fontsize=16, color='red', ha='center', va='center')
    ax.text(y_axis[0, 0]-x_label_offset, y_axis[0, 1]+y_label_offset, "y", fontsize=16, color='green', ha='center', va='center')
    
    # Axis labels for transformed coordinate system (matching the position of original coordinate system)
    ax.text(transformed_x_axis[0, 0]-x_label_offset, transformed_x_axis[0, 1]-y_label_offset, "x", fontsize=16, color='red', ha='center', va='center')
    ax.text(transformed_y_axis[0, 0]-x_label_offset, transformed_y_axis[0, 1]-y_label_offset, "y", fontsize=16, color='green', ha='center', va='center')

    # Add blue dot at position (2, 3) regarding coordinate system A
    ax.plot(2, 3, 'bo', markersize=14)

    # Increasing font size for the axis numbers
    ax.tick_params(axis='both', which='major', labelsize=14)
    # Even darker grid lines
    ax.grid(True, linestyle='--', which='both', linewidth=1, color='black')
    ax.axvline(x=0, color='gray', linestyle='--', linewidth=1)
    ax.axhline(y=0, color='gray', linestyle='--', linewidth=1)
    ax.set_xticks(np.arange(-2, 10, 1))
    ax.set_yticks(np.arange(-2, 7, 1))
    ax.set_xlim(-2, 9)
    ax.set_ylim(-2, 6)
    ax.set_aspect('equal', adjustable='box')

    
    
    plt.show()
    return fig, ax

# Run the function to plot with bold and larger labels
fig, ax = plot_transformed_system_with_matching_axis_labels()  # This function should return the fig object
fig.savefig("coordinate_systems.png", dpi=300, bbox_inches='tight', pad_inches=0)


```
